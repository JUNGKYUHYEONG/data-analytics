```python
# 실습 데이터 적재
data = pd.read_excel('data/인구수예제.xlsx')
data.head()
```

```python
#groupby() 실습
# 연도별 전체의 총인구수 구하기
# SELECT 연도, sum(총인구)
# FROM data
# GROUP BY 연도
data[data.연도==2013]['총인구'].sum()
```

```python
data.groupby('연도').총인구.sum()
```

```python
# 연도별 전체의 남자인구, 여자인구, 총인구 수 구하기
data.groupby('연도')[['남자인구','여자인구','총인구']].sum()
```

```python
# 자치구별로 평균 총인구수 구하기
```

```python
data.groupby(['도시','자치구'])[['총인구']].mean()
```

```python
# 도시/자치구별 평균 총인구수 구하기
```

```python
data.groupby(['자치구'])[['총인구']].mean()
```

```python
data.groupby(['도시'])[['총인구']].mean()
```

```python
# 실습 1. 도시/자치구별 평균 남자인구와 여자인구수 구하기
```

```python
data.groupby(['도시','자치구'])[['남자인구','여자인구']].mean()
```

```python
data.pivot_table(index=['도시','자치구'],
                values=['남자인구','여자인구'],aggfunc=['mean','std'])
```

```python
# 실습 2. 도시별, 연도별로 총인구수 출력
data.pivot_table(index=['도시','자치구'],
                values=['총인구'],aggfunc='mean')
```

```python
data.groupby(['도시','연도'])[['총인구']].sum().unstack(0)
```

```python
data.pivot_table(index='연도',columns='도시',
                values='총인구',aggfunc='sum')
```

```python
# pivot_table() 실습
# 실습. 도시/자치구 별로 남자인구의 평균을 구한 후, 남자인구가 가장 많은 도시/자치구 5개를 출력하시오. 
data.pivot_table(index=['도시','자치구'],values=['남자인구'],
                 aggfunc='mean').sort_values(by='남자인구',ascending=False).head(5)
```

```python
# 실습. 남자와 여자인구수 차이의 평균을 구한 후, 가장 많이 차이나는 도시/자치구 3개를 찾으시오. 
# Hint.abs() = 절대값을 구하는 함수
#남녀차이 컬럼을 추가
data['차이']=abs(data.남자인구-data.여자인구)
data.head()
```

```python
data.pivot_table(index=['도시','자치구'],values=['차이'],
                 aggfunc='mean').sort_values(by='차이',ascending=False).head(3)
```
